```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
from yellowbrick.cluster import KElbowVisualizer
from sklearn.metrics import silhouette_score
from sklearn.metrics import silhouette_score
```

```python
blueZoneDF = pd.read_csv("mumbai_Blue.csv")
redZoneDF = pd.read_csv("mumbai_Red.csv")
orangeZoneDF = pd.read_csv("mumbai_Orange.csv")
blueZoneDF_copy = blueZoneDF
redZoneDF_copy = redZoneDF
orangeZoneDF_copy = orangeZoneDF
```

```python
blueZoneDF.drop(blueZoneDF.columns[[0,3,4]],axis=1, inplace=True)
redZoneDF.drop(redZoneDF.columns[[0,3,4]],axis=1, inplace=True)
orangeZoneDF.drop(orangeZoneDF.columns[[0,3,4]],axis=1, inplace=True)
```

```python
blueZoneDF.head()
```

# For blue Zone

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(blueZoneDF)
y_kmeans = kmeans.predict(blueZoneDF)
```

```python
plt.scatter(blueZoneDF.iloc[:, 0], blueZoneDF.iloc[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

```python
print(silhouette_score(blueZoneDF, kmeans.labels_))
```

### Chosing 4 clusters as score is highest

```python
model = KMeans(random_state=0)
visualizer = KElbowVisualizer(model, k=(2,6), metric='silhouette', timings=False)
visualizer.fit(blueZoneDF)    
visualizer.poof()   
```

```python
kmeans.labels_
```

```python
blueZoneDF = blueZoneDF_copy.reset_index()
```

```python
clustersDf = pd.DataFrame(kmeans.labels_)
clustersDf.columns = ['cluster_predicted']
blueZoneDF = pd.concat([blueZoneDF, clustersDf], axis = 1).reset_index()
```

### Appending cluster data

```python
blueZoneDF.head()
```

```python
blueZoneDF.to_csv('blueZone_with_cluster.csv')
```
